## 11.Create a function to calculate porosity from RHOB using this equation as fraction. How many arguments can have default value for this function?
## Answer: Can have 1 default value


```python
def densityPorosity(rb, rf, rm):
    '''
    rb = bulk density (RHOB)
    rf = fluid density (default value of 1)
    rm = matrix density (limeston: 2.71 g/cc)
    '''
    por = (rm-rb)/(rm-rf)
    ron_por = round(por, 2)
    return ron_por
```
